```python
#Feature Engineering Time
def add_advanced_features(train):
    # 1. BMI and related
    train["BMI"] = train["Weight"] / ((train["Height"] / 100) ** 2)
    train["weight_height_ratio"] = train["Weight"] / train["Height"]

    # 2. BMR using Mifflin-St Jeor
    train["is_male"] = train["Sex"].map({"male": 1, "female": 0})
    train["BMR"] = (
        10 * train["Weight"] +
        6.25 * train["Height"] -
        5 * train["Age"] +
        train["is_male"].map({1: 5, 0: -161})
    )

    # 3. Calories per minute
    train["calories_per_minute"] = train["Calories"] / train["Duration"]
    train["age_group"] = pd.cut(train["Age"], bins=[10, 20, 30, 40, 50, 60, 100])
    train["avg_calories_by_age_group"] = train.groupby("age_group")["Calories"].transform("mean")

    # 4. Heart rate features
    train["resting_hr_diff"] = train["Heart_Rate"] - 70
    train["hr_duration_product"] = train["Heart_Rate"] * train["Duration"]
    train["hr_per_kg"] = train["Heart_Rate"] / train["Weight"]

    # 5. Body temperature features
    train["body_temp_deviation"] = train["Body_Temp"] - 37
    train["temp_hr_product"] = train["Body_Temp"] * train["Heart_Rate"]
    train["temp_duration_interaction"] = train["Body_Temp"] * train["Duration"]

    # 6. Interaction features
    train["age_weight_product"] = train["Age"] * train["Weight"]
    train["duration_weight"] = train["Duration"] * train["Weight"]
    train["age_duration_product"] = train["Age"] * train["Duration"]
    train["temp_weight_ratio"] = train["Body_Temp"] / train["Weight"]

    # 7. Polynomial / squared features
    train["weight_squared"] = train["Weight"] ** 2
    train["duration_squared"] = train["Duration"] ** 2
    train["hr_squared"] = train["Heart_Rate"] ** 2
    train["bmi_hr_product"] = train["BMI"] * train["Heart_Rate"]

    # 8. Standard scaling for select columns
    scale_cols = ["Age", "Height", "Weight", "Duration", "Heart_Rate", "Body_Temp", "BMI", "BMR"]
    scaler = StandardScaler()
    train[scale_cols] = scaler.fit_transform(train[scale_cols])

    # Convert 'Sex' to 0/1
    train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})  # adjust if your dataset uses different strings

    # Convert any remaining category columns to numeric
    for col in train.select_dtypes(['category']).columns:
      train[col] = train[col].cat.codes
    train = train.apply(pd.to_numeric, errors='coerce')

    return train
```

```python
add_advanced_features(train)
```

```python
#Selecting which model is best to use
X = train.drop('Calories', axis=1)
y = train['Calories']

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)

xgb = XGBRegressor(objective="reg:squarederror", n_estimators=200, learning_rate=0.05, max_depth=3, subsample=0.8, colsample_bytree=0.8, gamma=0.1, reg_alpha=0.1, reg_lambda=1.5, random_state=42)
xgb_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2]
}

xgb.fit(X_train, Y_train)

y_pred = xgb.predict(X_test)

mae = mean_absolute_error(Y_test, y_pred)
mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```

```python
#Making The Machine Learning Pipeline to never have any dataleak making the training have less bias
X = train.drop('Calories', axis=1)
y = train['Calories']

categorical_features = ['Sex']
numerical_features = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']

# 2. Preprocessor
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

# 3. Final XGB model with your best parameters
xgb = XGBRegressor(
    objective="reg:squarederror",
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.1,
    reg_alpha=0.1,
    reg_lambda=1.5,
    random_state=42
)

# 4. Full pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', xgb)
])

# 5. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 6. Fit model
pipeline.fit(X_train, y_train)

# 7. Predict and evaluate
y_pred = pipeline.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R^2 Score: {r2:.2f}")
print(y_pred)
```

```python
#Making The Machine Learning Pipeline to never have any dataleak making the training have less bias
X = train.drop('Calories', axis=1)
y = train['Calories']

categorical_features = ['Sex']
numerical_features = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']

# 2. Preprocessor
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

base_estimator = DecisionTreeRegressor(max_depth=5, min_samples_split=5)

# 3. Final XGB model with your best parameters
ada = AdaBoostRegressor(
    estimator=base_estimator, 
    random_state=42,
    n_estimators=200,
    learning_rate=0.05
)

# 4. Full pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', ada)
])

# 5. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 6. Fit model
pipeline.fit(X_train, y_train)

# 7. Predict and evaluate
y_pred = pipeline.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R^2 Score: {r2:.2f}")
print(y_pred)
```
